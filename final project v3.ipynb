```python
#Data retrieving

#Dependencies
from iexfinance.stocks import get_historical_data # Python module to get stock data from IEX Cloud and IEX API 1.0 (https://pypi.org/project/iexfinance/)
from datetime import datetime
import pandas as pd
```

```python
start = datetime(2019, 3, 1)
end = datetime.today()

company_stock_symbols = ["TSLA", "GOOGL", "NFLX"]
company_names = ["Tesla, Inc.", "Google LLC", "Netflix, Inc."]

stock_df = pd.DataFrame()

for item in range(len(company_stock_symbols)):
    temp_df = get_historical_data(str(company_stock_symbols[item]), start, end, output_format='pandas')
    temp_df ["Company Name"] = str(company_names[item])
    stock_df = stock_df.append(temp_df, sort=False)
  
```

```python
# Create columns for dataframe
tweets_table = {
                'Tweet Number': '',
                'PersonTweeted': '',
                'Tweet': '',
                'Compound Score': '',
                'Positive Score': '',
                'Negative Score': '',
                'Neutral Score': ''}

# Create list of news accounts
target_terms = ['@elonmusk', '@sundarpichai', '@reedhastings']

# Variable to hold list of the compound sentiment
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
tweet_number = []
person_tweeted = []
text = []
created =[]
retweet = []
likes = [] 

# Loop through all target users
for target in target_terms:
        
    # Run search around each tweet(Upto max of 200 tweets at once)
    screen_name = target
    public_tweets = api.user_timeline(target, count=200,result_type ="recent")
    
    # Index to see how many tweets for each news source
    index = 0
        
    # Loop through all tweets    
    for tweet in public_tweets:
        
        news_organization = target
            
        # Run vader analysis on each tweet        
        scores = analyzer.polarity_scores(tweet['text'])
        compound = scores['compound']
        pos = scores['pos']
        neu = scores['neu']
        neg = scores['neg']
            
        # Append values to empty lists        
        compound_list.append(compound)
        positive_list.append(pos)
        negative_list.append(neg)
        neutral_list.append(neu)
        text.append(tweet['text'])
        tweet_number.append(index)
        person_tweeted.append(target)
        tweet_date = datetime.datetime.strptime(tweet["created_at"], "%a %b %d %H:%M:%S %z %Y")
        convert_time = datetime.datetime.strftime(tweet_date, '%y-%m-%d')
        created.append(convert_time)
        retweet.append(tweet["retweet_count"])
        likes.append(tweet["favorite_count"])      
        
        index = index + 1
        
```

```python
#Create dataframe from the lists appended above 
tweets_table = {
                'Person_Tweeted' : person_tweeted,
                'Tweet Number' : tweet_number,
                'Tweet' : text,
                'Compound Score' : compound_list,
                'Positive Score': positive_list,
                'Negative Score': negative_list,
                'Neutral Score': neutral_list,    
                'Date': created,
                'Retweet' : retweet,
                'Likes' : likes}

# tweets_df = pd.DataFrame(tweets_table)
# tweets_df
```

```python
# Init
newsapi = NewsApiClient(api_key='3e89e595568a4616a16181234c5c4e25')

queries = ["google", "tesla", "netflix"]

authors = []
titles = []
descriptions = []
sources = []
publish_dates = []
topics = []

for query in range(len(queries)):
    all_articles = newsapi.get_everything(q=queries[query],
                                          #from_param='2019-03-01',
                                          from_param=str(datetime.today().year)+"-"+str(datetime.today().month - 1)+"-"+str(datetime.today().day + 1),
                                          to=str(datetime.today()),
                                          language='en',
                                          sort_by='relevancy',
                                          page=5)    
    for num in range(len(all_articles['articles'])):
        topics.append(queries[query])
        authors.append(all_articles['articles'][num]['author'])
        titles.append(all_articles['articles'][num]['title'])
        descriptions.append(all_articles['articles'][num]['description'])
        sources.append(all_articles['articles'][num]['url'])
        publish_dates.append(str(all_articles['articles'][num]['publishedAt']).split("T")[0])
            
            

```

```python
news_dict = {"Topic": topics,
             "Author": authors,
             "Title": titles,
             "Description": descriptions,
             "Source": sources,
             "Date": publish_dates}

news_df = pd.DataFrame(news_dict)
```

```python
news_df["Company Name"] = ""
news_df.loc[news_df["Topic"] == "tesla", "Company Name"] = "Tesla, Inc."
news_df.loc[news_df["Topic"] == "google", "Company Name"] = "Google LLC"
news_df.loc[news_df["Topic"] == "netflix", "Company Name"] = "Netflix, Inc."
```

```python
#News sentiment analysis
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

# Run vader analysis on each tweet

for index, row in news_df.iterrows():
    scores = analyzer.polarity_scores(row['Description'])
    compound = scores['compound']
    pos = scores['pos']
    neu = scores['neu']
    neg = scores['neg']

    # Append values to empty lists

    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)
```
